```python
import pandas as pd
import numpy as np

# Step 1: Generate student IDs and random grades
student_ids = pd.Series([f'ST_{i+1}' for i in range(10)]) 
grades = pd.Series(np.random.randint(0, 101, size=10)) 

# Display original student IDs and grades
print("Original Student IDs and Grades:")
for student_id, grade in zip(student_ids, grades):
    print(f"{student_id}: {grade}")

# Step 2: Add a 10-point boost to all grades
grades += 10

# Step 3: Replace all grades below 50 with 0
grades = grades.where(grades >= 50, 0)

# Display updated student IDs and grades
print("\nUpdated Student IDs and Grades:")
for student_id, grade in zip(student_ids, grades):
    print(f"{student_id}: {grade}")
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("heart.csv")

# Step 1: Print the column names and number of rows in the data
print("Column Names:")
print(data.columns.tolist())
print("\nNumber of Rows:", len(data))

# Step 2: Get a row number from the user
try:
    row_number = int(input("\nEnter a row number (0-indexed): "))
    if row_number < 0 or row_number >= len(data):
        raise ValueError("Row number out of range.")
except ValueError as e:
    print("Invalid input for row number:", e)
    exit()

# Step 3: Get a column name from the user
column_name = input("Enter a column name: ")

# Step 4: Print the specified row, column, and the value at the intersection
try:
    # Print the entire row
    print("\nRow:")
    print(data.iloc[row_number])

    # Print the specified column
    print("\nColumn:")
    print(data[column_name])

    # Print the value at the intersection of the specified row and column
    print(f"\nValue at Row {row_number}, Column '{column_name}':", data.at[row_number, column_name])
except KeyError:
    print("Invalid column name.")
except IndexError:
    print("Row number out of range.")
```

```python
import pandas as pd

df = pd.read_csv("Team Summaries.csv")
display(df.head())
display(df['w'].value_counts())
print("Descriptive:")
print(df[["w", "l", "pw", "pl"]].describe())

print("\nAverage win per season:")
print(df.groupby(["season","team"])["w"].mean())

def calculate_best_team(df):

    df["Weighted_Score"] = (df["w"] * 0.5) + (df["mov"] * 0.25) + (df["sos"] * 0.25)
    best_team = df.loc[df["Weighted_Score"].idxmax()]
    return best_team

best_team = calculate_best_team(df.copy()) 
print("\nWeighted score, the 'best' team is:", best_team)
```
