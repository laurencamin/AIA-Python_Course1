## 6.1 GRAPHS AND TREES (NETWORKX) (Group)

Write a program that will do the following:

1. Create a graph based on user input
2. Take number of nodes from user and ask user to enter edges (you can get creative in taking inputs)
3. (Optional) Get labels of each node
4. Use networkx to create a graph / tree as required by the user.
5. Display the graph


```python

import networkx as nx
import matplotlib.pyplot as plt

def create_graph():
    # Step 1: Get the number of nodes
    num_nodes = int(input("Enter the number of nodes: "))
    
    # Step 2: Get optional labels for each node
    labels = {}
    for i in range(num_nodes):
        label = input(f"Enter label for node {i+1} (or press Enter to skip): ")
        if label:
            labels[i] = label

    # Step 3: Get edges from the user
    edges = []
    print("Enter edges in the format 'node1 node2'. Type 'done' when finished.")
    while True:
        edge_input = input("Enter edge: ")
        if edge_input.lower() == 'done':
            break
        try:
            node1, node2 = map(int, edge_input.split())
            edges.append((node1 - 1, node2 - 1))  # Adjusting for 0-based index
        except ValueError:
            print("Invalid input. Please enter two node numbers separated by a space.")

    # Step 4: Create the graph
    G = nx.Graph()
    G.add_nodes_from(range(num_nodes))
    G.add_edges_from(edges)

    # Step 5: Draw the graph
    pos = nx.spring_layout(G)  # positions for all nodes
    nx.draw(G, pos, with_labels=True, labels=labels, node_color='blue', node_size=700, font_size=10, font_color='black')
    
    # Step 6: Display the graph
    plt.title("Graph Visualization")
    plt.show()

if __name__ == "__main__":
    create_graph()
```

## 6.2 GRAPHS AND TREES (Group)

Write a program that will do the following:

1. Use the Class Graph to create a random graph
2. Get number of nodes and number of edges from the user.
3. Randomly assign edges to the graph (e > number of nodes)
4. Display the resulting graph.

```python
import random
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    def __init__(self, num_nodes):
        self.num_nodes = num_nodes
        self.graph = nx.Graph()

    def add_random_edges(self, num_edges):
        # Ensure that the number of edges is greater than the number of nodes
        if num_edges < self.num_nodes:
            raise ValueError("Number of edges must be greater than or equal to the number of nodes.")

        # Add nodes to the graph
        self.graph.add_nodes_from(range(self.num_nodes))

        # Randomly add edges
        edges_added = 0
        while edges_added < num_edges:
            node1 = random.randint(0, self.num_nodes - 1)
            node2 = random.randint(0, self.num_nodes - 1)

            # Avoid self-loops and duplicate edges
            if node1 != node2 and not self.graph.has_edge(node1, node2):
                self.graph.add_edge(node1, node2)
                edges_added += 1

    def display_graph(self):
        pos = nx.spring_layout(self.graph)  # positions for all nodes
        nx.draw(self.graph, pos, with_labels=True, node_color='blue', node_size=700, font_size=10, font_color='black')
        plt.title("Random Graph Visualization")
        plt.show()

def main():
    # Get user input for number of nodes and edges
    num_nodes = int(input("Enter the number of nodes: "))
    num_edges = int(input("Enter the number of edges (must be greater than number of nodes): "))

    # Create a graph and add random edges
    graph = Graph(num_nodes)
    try:
        graph.add_random_edges(num_edges)
        graph.display_graph()
    except ValueError as e:
        print(e)

if __name__ == "__main__":
    main()
    
```

```python
from collections import defaultdict
import networkx as nx

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        if self.root is None:
            self.root = TreeNode(value)
        else:
            self._insert_recursively(self.root, value)

    def _insert_recursively(self, node, value):
        if value < node.value:
            if node.left is None:
                node.left = TreeNode(value)
            else:
                self._insert_recursively(node.left, value)
        else:
            if node.right is None:
                node.right = TreeNode(value)
            else:
                self._insert_recursively(node.right, value)

    def is_binary_tree(self):
        return self._is_binary_tree(self.root)

    def _is_binary_tree(self, node):
        if node is None:
            return True
        # A binary tree node can have at most two children
        if (node.left is not None and node.right is not None):
            return False
        return self._is_binary_tree(node.left) and self._is_binary_tree(node.right)

def main():
    # Create a tree
    tree = Tree()

    # Get user input for generating a tree
    values = input("Enter values to insert into the tree (comma-separated): ")
    values = [int(value.strip()) for value in values.split(",")]

    # Insert values into the tree
    for value in values:
        tree.insert(value)

    # Check if the tree is a binary tree
    if tree.is_binary_tree():
        print("The tree is a binary tree.")
    else:
        print("The tree is NOT a binary tree.")

if __name__ == "__main__":
    main()
```

```python
from collections import defaultdict
import networkx as nx
import matplotlib.pyplot as plt

class Tree:
    def __init__(self):
        self.tree = defaultdict(list)
    
    def __str__(self):
        s = ""
        for i in self.tree:
            s += f'\n {i} -> {self.tree[i]}'
        return s
 
    def addEdge(self, u, v):
        self.tree[u].append(v)
    
    def plotTree(self):
        plot = nx.DiGraph()
        elist = []
        for i in self.tree:
            for j in self.tree[i]:
                elist.append((i, j))
        plot.add_edges_from(elist)
        nx.draw(plot, with_labels=True)
        plt.show()

    def post_order_traversal(self, start_node):
        visited = set()  # To keep track of visited nodes
        post_order = []  # To store the post-order traversal

        def dfs(node):
            visited.add(node)  # Mark the node as visited
            for neighbor in self.tree[node]:  # Use self.tree to access neighbors
                if neighbor not in visited:
                    dfs(neighbor)  # Recursively visit the neighbor
            post_order.append(node)  # Add the node to post-order after visiting its neighbors

        dfs(start_node)  # Start DFS from the given start node
        return post_order

def main():
    t = Tree()
    t.addEdge('f', 'b')
    t.addEdge('f', 'g')
    t.addEdge('b', 'a')
    t.addEdge('b', 'd')
    t.addEdge('d', 'c')
    t.addEdge('d', 'e')
    t.addEdge('g', 'i')
    t.addEdge('i', 'h')

    t.plotTree()
    print(t)
    print("POSTORDER TRAVERSAL")
    result = t.post_order_traversal('f')
    print(result)

if __name__ == "__main__":
    main()
```

3. Learn algorithms for Depth First and Breadth first Searching
4. Understand the necessity of Graph searching.

```python
"""
Add comments for you own reference
"""

from collections import defaultdict
import networkx as nx


class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
 
    def addEdge(self,u,v):
        self.graph[u].append(v)
    
    def plotGraph(self):
        plot = nx.DiGraph()
        elist = []
        for i in self.graph:
            for j in self.graph[i]:
                elist.append((i,j))
        plot.add_edges_from(elist)
        nx.draw(plot, with_labels = True)
    
    
    def BFS(self, s, g):
        visited_queue = []
        BFS_queue = []
 
        BFS_queue.append(s)
        visited_queue.append(s)
        if s == g:
            print("Goal found!", visited_queue)
            return
            
        while BFS_queue:
            s = BFS_queue.pop(0)
 
            for i in self.graph[s]:
                if i not in visited_queue:
                    BFS_queue.append(i)
                    visited_queue.append(i)
                    if i == g:
                        print("Goal found!", visited_queue)
                        return
                    
                    
    def DFS(self, s, g):
        visited_list = []
        DFS_stack = []
 
        DFS_stack.append(s)
        visited_list.append(s)
        if s == g:
            print("Goal found!", visited_list)
            return
    
        while DFS_stack:
            s = DFS_stack.pop(-1)

            for i in self.graph[s]:
                if i not in visited_list:
                    DFS_stack.append(i)
                    visited_list.append(i)
                    if i == g:
                        print("Goal found!", visited_list)
                        return
```

## 6.5 SEARCHING (Group)

Write a program that will do the following:

1. Implement depth first search and breadth first search.
2. Implement a class Node.
3. Node class should have attributes value, and parent.
4. Graph class should **not** inherit the node class.
5. Create a graph object and use the add edge function to create a graph of nodes.  
6. Create at least 5 or 6 nodes.
7. Perform DFS and print the results of DFS and BFS.
8. As an added challenge, create a function to keep track of the parents of each node.
9. If you create the function, try printing these parents.


```python
from collections import defaultdict, deque

class Node:
    def __init__(self, value):
        self.value = value
        self.parent = None

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.nodes = {}

    def add_node(self, value):
        if value not in self.nodes:
            self.nodes[value] = Node(value)

    def add_edge(self, u, v):
        self.add_node(u)
        self.add_node(v)
        self.graph[u].append(v)
        self.graph[v].append(u)  # Assuming an undirected graph

    def dfs(self, start):
        visited = set()
        parents = {}
        result = []

        def dfs_recursive(node):
            visited.add(node)
            result.append(node)
            for neighbor in self.graph[node]:
                if neighbor not in visited:
                    parents[neighbor] = node  # Track the parent
                    dfs_recursive(neighbor)

        dfs_recursive(start)
        return result, parents

    def bfs(self, start):
        visited = set()
        queue = deque([start])
        visited.add(start)
        parents = {start: None}
        result = []

        while queue:
            node = queue.popleft()
            result.append(node)
            for neighbor in self.graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
                    parents[neighbor] = node  # Track the parent

        return result, parents

def main():
    # Create a graph object
    g = Graph()

    # Add edges to create a graph of nodes
    g.add_edge('A', 'B')
    g.add_edge('A', 'C')
    g.add_edge('B', 'D')
    g.add_edge('B', 'E')
    g.add_edge('C', 'F')
    g.add_edge('E', 'F')

    # Perform DFS and BFS
    print("DFS Traversal:")
    dfs_result, dfs_parents = g.dfs('A')
    print(dfs_result)
    print("Parents in DFS:", dfs_parents)

    print("\nBFS Traversal:")
    bfs_result, bfs_parents = g.bfs('A')
    print(bfs_result)
    print("Parents in BFS:", bfs_parents)

if __name__ == "__main__":
    main()
        
```
