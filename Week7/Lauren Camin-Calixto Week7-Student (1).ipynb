Write a program that will do the following:

3. Retrive the number by reading the file
4. Take the number from the file and multiply it by another random number.
5. Append the file with the multiplication of the two numbers.

```python
import random

def main():
    initial_number = random.randint(1, 100)  #generate a random number between 1 and 100
    with open("Workshop.txt", 'w') as file:
        file.write(str(initial_number) + '\n')
        
    file = open('Workshop.txt', 'r') #open and read file
    print(file.read())

    for i in range(5):  
        with open('Workshop.txt', 'r') as file:
            number = int(file.readline().strip())  #read the first line and convert to an integer

        random_multiplier = random.randint(1, 10)  
        result = number * random_multiplier

        with open('Workshop.txt', 'a') as file:
            file.write(f'{result}\n')  

        #print numbers for verification
        print(f'Original number: {number}, Multiplier: {random_multiplier}, Result: {result}')

main()
        
    
```

## 7.2 CURSORS (Group)

Write a program that will do the following:

1. Use the file 'Workshop.txt'
2. Read the file first to get the total number of characters in file.
3. Use seek and tell to create a new file 'Workshop_rev.txt' that is a copy of the file in reverse order.
4. Read and print 'Workshop_rev.txt'

```python

with open('Workshop.txt', 'r') as file:
    content = file.read()
    total_characters = len(content)
    print(f'Total number of characters in Workshop.txt: {total_characters}')

with open('Workshop.txt', 'r') as file:
    with open('Workshop_rev.txt', 'w') as rev_file:
        file.seek(0, 2)  # Move the cursor to the end of the file
        position = file.tell()  # Get the current position (end of file)
        
        while position > 0:
            position -= 1  # Move one character back
            file.seek(position)  # Move the cursor to the current position
            char = file.read(1)  # Read one character
            rev_file.write(char)  # Write the character to the new file


with open('Workshop_rev.txt', 'r') as rev_file:
    reversed_content = rev_file.read()
    print('Content of Workshop_rev.txt:')
    print(reversed_content)
```

```python
"""
Add comments for you own reference
"""

import csv

with open('grades.csv', mode = 'w', newline='') as grades:
        writer = csv.writer(grades)
        writer.writerow([1, 'Tom', 98])
        writer.writerow([2, 'Matt', 75])
        writer.writerow([3, 'Bob', 62])
        writer.writerow([4, 'Kev', 45])
        writer.writerow([5, 'Pam', 23])
        writer.writerow([6, 'Sam', 100])
        
with open('grades.csv', 'r', newline='') as grades:
    print(f'{"ID":<4}{"Name":<7}{"Score"}')
    reader = csv.reader(grades)
    for record in reader:
        student_id, name, score = record
        print(f'{student_id:<4}{name:<7}{score}')
```

```python
"""
Add comments for you own reference
"""

import io

def main():
    name = input("Enter file name: ")
    try:
        file = open(name,'r')
        print(file.read())
        #file.write("HEY")
        file.close()
    except FileNotFoundError:
        print("No such file exists !")
    except PermissionError:
        print("You do not have access here")
    except io.UnsupportedOperation:
        print("File opened in read mode !")
main()
```

```python
import csv

class Athlete:
    def __init__(self, *args):
        self.__attributes = {}
        for X in range(len(args[0])):
            self.__attributes[args[0][X]] = args[1][X]
        
    def get_age(self):
        try:
            age = int(self.__attributes['Age'])
            if age > 0:
                return age
        except (ValueError, KeyError):
            return 0

    def get_height(self):
        try:
            height = int(self.__attributes['Height'])
            if height > 0:
                return height
        except (ValueError, KeyError):
            return 0

    def get_weight(self):
        try:
            weight = int(self.__attributes['Weight'])
            if weight > 0:
                return weight
        except (ValueError, KeyError):
            return 0

athlete_list = []
age_total = 0
height_total = 0
weight_total = 0
height_list = []

# Read the CSV file and create Athlete objects
with open('athlete_events_short.csv', 'r', newline='') as athletes:
    reader = csv.reader(athletes)
    header = next(reader)  # Read the header

    for line in reader:
        new_athlete = Athlete(header, line)
        athlete_list.append(new_athlete)

# Calculate total age, height, and weight
for athlete in athlete_list:
    age_total += athlete.get_age()
    height = athlete.get_height()
    weight = athlete.get_weight()
    height_total += height
    weight_total += weight
    height_list.append(height)

# Calculate averages and find the tallest athlete
if athlete_list:  # Check if the list is not empty
    average_age = age_total / len(athlete_list)
    average_height = height_total / len(athlete_list)
    average_weight = weight_total / len(athlete_list)
    tallest_athlete = max(height_list)

    print("Average age in years for all Olympians =", average_age)
    print("Average height for all Olympians =", average_height)
    print("Average weight for all Olympians =", average_weight)
    print("The tallest Olympian is", tallest_athlete)
else:
    print("No athletes in the list.")
```
