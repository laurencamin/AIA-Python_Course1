## 8.3 STATISTICS (Group)

Write a program that will do the following:

1. Get a number n from User (10-30)
2. Print the statistic information for every nth row of the 'test.csv'
    - Mean
    - Median
    - Sum
    - SD
    - Variation


```python
"""
Add comments for you own reference
"""

import numpy as np

num = n = int(input("Enter number between 10 -30: "))
data = np.loadtxt("test.csv",delimiter=",", dtype=int)


while n < len(data):
    print(f"Mean of {n}th -")
    print(data[n].mean())
    print(f"Median of {n}th -")
    print(np.median(data[n]))
    print(f"Sum of {n}th -")
    print(np.sum(data[n]))
    print(f"Standard Deviation of {n}th -")
    print(np.std(data[n]))
    print(f"Variation of {n}th -")
    print(np.var(data[n]))
    print()
    n+= num
    
```

## 8.5 COPY VS VIEW (Follow)

**Learning Objectives:**

1. Understand the difference between a copy and a view.
2. Understand how the original array is affected when deling with views.
3. Understand how the original array is affected when deling with copies.
4. Learn to manipulate data using copies and views

```python
#https://data-flair.training/blogs/numpy-copy-view/
"""
Add comments for you own reference
"""

arr = np.array([1,2,3,4,5])
print("Original Before -")
print(arr)
x = arr.view()
x[0] = 9
print("Original for view -")
print(arr)
print("View -")
print(x)
y = arr.copy()
y[1] = 8
print("Original for copy -")
print(arr)
print("Copy -")
print(y)
```

```python
"""
Add comments for you own reference
"""

data = np.loadtxt("test.csv",delimiter=",", dtype=int)
data = data[5:10]

print("Original Before -")
print(data)
x = data.view()
x[0] = 9
print("Original for view -")
print(data)
print("View -")
print(x)
y = data.copy()
y[1] = 8
print("Original for copy -")
print(data)
print("Copy -")
print(y)
```

```python
import numpy as np
data = np.loadtxt("test.csv",delimiter=",", dtype=int)
data = data[0:5, :]
data_view = data.view()
data_copy = data.copy()

for i in range(0,5,2):
    data_view[i] = 5

for i in range(1,5,2):
    data_copy[i] = 10
print(data)
print(data_copy)


data_view += data_copy
data_view = data_view + data_copy

print("\n----------------------------------")
print("data view memory location: ", data_view[0].data)
print(data_view)
print("\ndata memory location: ", data[0].data)
print(data)
print("\ndata copy memory location: ", data_copy[0].data)
print(data_copy)
```

## 8.6 MODIFICATION (Follow)

**Learning Objectives:**

1. Understand how to modify an array.
2. Learn modification functions.
3. Understand the use of array modification

```python
"""
Add comments for you own reference
"""

import numpy as np
arr = np.array([1,2,3,4,5,6])
arr.resize(2,3)
print('arr = \n',arr,"\n")

x = arr.reshape(3,2)
print('x = \n', x,"\n")
#print('arr = \n', arr, '\n')

y = x.flatten() #copy
z = x.ravel() #view
y[0] = 99
z[0] = 88
print('x = \n', x,"\n")
print('y = \n', y,"\n")
print('z = \n', z,"\n")
print(x.transpose(),"\n")
```

```python
"""
Add comments for you own reference
"""

data = np.loadtxt("test.csv",delimiter=",", dtype=str).astype(int)
data = data[25:30]

data.resize(25,2)
print('data = \n', data,"\n")
x = data.reshape(10,5)
print('x = \n', x,"\n")
y = x.flatten()
z = x.ravel()
y[0] = 99
z[0] = 88
print('x = \n', x,"\n")
print('y = \n', y,"\n")
print('z = \n', z,"\n")
print(x.transpose(),"\n")
```

## 8.6 MODIFICATION (Group)

Write a program that will do the following:

1. Read 'test.csv' 
2. Slice a 5 X 8 grid from the array (X)
3. Make a copy and transpose (X)
4. Reshape both arrays to (4,10)
5. Multiply the arrays and display the Output
6. Try to display each step

```python
"""
Add comments for you own reference
"""

import numpy as np
data = np.loadtxt("test.csv",delimiter=",", dtype=int)

n = int(input("Enter Number: "))

dataX = data[n:n+5,1:9]

print(dataX.shape)

dataX_copy = dataX.copy()

print("\n COPY:")
print(dataX_copy)

dataY = dataX_copy.transpose()

print("\n TRANSPOSE")
print(dataY)

X = dataX.reshape(4,10)
Y = dataY.reshape(4,10)

print("\n RESHAPED:")
print(X)
print("\n RESHAPED:")
print(Y)

ans = X * Y

print("\nANSWER")
print(ans)
```
