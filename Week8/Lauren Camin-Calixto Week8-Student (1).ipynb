```python
import numpy as np

def get_statistics_for_nth_rows(file_path, n):
    # Load the CSV file into a NumPy array
    data = np.loadtxt("test.csv", delimiter=',', dtype= int)  

    # Check if n is within the valid range
    if n < 10 or n > 30:
        print("Please enter a number between 10 and 30.")
        return

    # Iterate over every nth row
    for i in range(0, data.shape[0], n):
        row = data[i]

        # Calculate statistics
        mean = np.mean(row)
        median = np.median(row)
        total_sum = np.sum(row)
        std_dev = np.std(row)
        variance = np.var(row)

        # Print statistics
        print(f"Statistics for row {i}:")
        print(f"Mean: {mean}")
        print(f"Median: {median}")
        print(f"Sum: {total_sum}")
        print(f"Standard Deviation: {std_dev}")
        print(f"Variance: {variance}")
        print("-" * 40)

# Get user input
try:
    n = int(input("Enter a number between 10 and 30: "))
    file_path = 'test.csv' 
    get_statistics_for_nth_rows(file_path, n)
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

## 8.5 COPY VS VIEW (Follow)

**Learning Objectives:**

1. Understand the difference between a copy and a view.
2. Understand how the original array is affected when deling with views.
3. Understand how the original array is affected when deling with copies.
4. Learn to manipulate data using copies and views

```python
"""
Add comments for you own reference
"""

arr = np.array([1,2,3,4,5])
print("Original Before -")
print(arr)
x = arr.view()
x[0] = 9
print("Original for view -")
print(arr)
print("View -")
print(x)
y = arr.copy()
y[1] = 8
print("Original for copy -")
print(arr)
print("Copy -")
print(y)
```

```python
"""
Add comments for you own reference
"""

data = np.loadtxt("test.csv",delimiter=",", dtype=int)
data = data[5:10]

print("Original Before -")
print(data)
x = data.view()
x[0] = 9
print("Original for view -")
print(data)
print("View -")
print(x)
y = data.copy()
y[1] = 8
print("Original for copy -")
print(data)
print("Copy -")
print(y)
```

**Evaluation:**

After running this program, think about and answer these questions:

1. What does the copy() function do? It creates an array with its own separate data. Modifications to the copy do not affect the original array, and vice versa.
2. What does the view() function do? A view is a new array object that looks at the same data as the original array. Changes made to a view will affect the original array, and changes to the original will affect the view.
3. How is the original array affected by manipulating views / copies? Modifications to the copy do not affect the original array, and vice versa. Changes made to a view will affect the original array, and changes to the original will affect the view.
4. What is the best use case for views? Use views when you want to access or modify data without the overhead of copying, such as when working with large datasets.
5. What is the best use case for copies?  Use copies when you need to modify an array without altering the original data.

## 8.5 COPY VS VIEW (Group)

Write a program that will do the following:

1. Read 'test.csv' 
2. Make a copy and a view of the data
3. Convert even rows of view to all 5
4. Convert odd rows of copy to all 10
5. Add Copy to view
6. Display all three Arrays

```python
import numpy as np
import csv 

data = np.loadtxt("test.csv",delimiter=",", dtype=int)
x = data.view()
y = data.copy()
```

```python

import numpy as np

def main():
    # Load the CSV file into a NumPy array
    file_path = 'test.csv' 
    data = np.genfromtxt(file_path, delimiter=',', dtype=int)  

    # Create a view and a copy of the data
    x = data.view()
    y = data.copy()

    # Convert even rows of the view to all 5
    x[::2] = 5  

    # Convert odd rows of the copy to all 10
    y[1::2] = 10  

    # Add the copy to the view
    result_data = x + y

    # Display all three arrays
    print("Original Data:")
    print(data)
    print("\nView Data (Even rows set to 5):")
    print(x)
    print("\nCopy Data (Odd rows set to 10):")
    print(y)
    print("\nResult Data (View + Copy):")
    print(result_data)

if __name__ == "__main__":
    main()
```

```python
"""
Add comments for you own reference
"""

import numpy as np
arr = np.array([1,2,3,4,5,6])
print('arr = \n',arr,"\n")
arr.resize(2,3) #view
print('arr = \n',arr,"\n")
x = arr.reshape(3,2) #copy
print('x = \n', x,"\n")

y = x.flatten()  #copy
z = x.ravel()    #view
y[0] = 99
z[0] = 88
print('x = \n', x,"\n")
print('y = \n', y,"\n")
print('z = \n', z,"\n")
print(x.transpose(),"\n")
print(x)
```

## 8.6 MODIFICATION (Group)

Write a program that will do the following:

1. Read 'test.csv' 
2. Slice a 5 X 8 grid from the array (X)
3. Make a copy and transpose (X)
4. Reshape both arrays to (4,10)
5. Multiply the arrays and display the Output
6. Try to display each step

```python
import numpy as np

def main():
    # Step 1: Load the CSV file into a NumPy array
    file_path = 'test.csv'  # Replace with your CSV file path
    data = np.genfromtxt(file_path, delimiter=',', dtype=int)  

    # Step 2: Slice a 5x8 grid from the array
    # Assuming the original data has at least 5 rows and 8 columns
    X = data[:5, :8]
    print("Step 2: 5x8 Grid (X):")
    print(X)

    # Step 3: Make a copy and transpose (X)
    X_copy = X.copy()
    X_transpose = X_copy.T
    print("\nStep 3: Transposed Copy of X:")
    print(X_transpose)

    # Step 4: Reshape both arrays to (4, 10)
    X_reshaped = X.reshape(4, 10)
    X_transpose_reshaped = X_transpose.reshape(4, 10)
    print("\nStep 4: Reshaped X (4, 10):")
    print(X_reshaped)
    print("\nStep 4: Reshaped Transposed X (4, 10):")
    print(X_transpose_reshaped)

    # Step 5: Multiply the arrays and display the output
    result = X_reshaped * X_transpose_reshaped  # Using matrix multiplication
    print("\nStep 5: Result of Multiplication:")
    print(result)

if __name__ == "__main__":
    main()
```
