## 5.5 Super versus Object Init (Follow):  
**Learning Objectives:**

1. Understand how super and a regular object init differ.
2. Understand when it is appropriate to use super versus regular object init.
3. Understand what the diamond inheritance pattern is.

```python
class A:
    def __init__(self, num):
        print("In A")
        self.data = num
       
    def __str__(self):
        return f'A data = {self.data}'

class B(A): 
    def __init__(self,num):
        print("In B")
        super().__init__(num +1)
        
    def __str__(self):
        return f'B data = {self.data}'
    
class C(A):
    def __init__(self, num):
        print("In C")
        super().__init__(num + 2)
 
    def __str__(self):
        return f'C data = {self.data}'
    
class D(B, C):
    def __init__(self, num):
        print("In D")
        super().__init__(num + 3)

    def __str__(self):
        return f'D data = {self.data}'

objA = A(1)
print(f'A: {objA} \n')

objB = B(2)
print(f'B: {objB} \n')

objC = C(3)
print(f'C: {objC} \n')

objD = D(4)
print(f'D: {objD} \n')

print(D.__mro__)
```

```python
class A():
    def __init__(self):
        print('Running A.__init__')
        super(A, self).__init__()

class B(A):
    def __init__(self):
        print('Running B.__init__')        
        super(B, self).__init__()

class C(A):
    def __init__(self):
        print('Running C.__init__')
        super(C, self).__init__()

class D(B,C):
    def __init__(self):
        print('Running D.__init__')
        super(B, self).__init__()

the_object = D()
D.mro()
```

```python
from collections import deque

class Nonprofit:
    def __init__(self, name, location, issue):
        self.__name = name
        self.__location = location
        self.__issue = issue

    def set_name(self, name):
        self.__name = name

    def set_location(self, location):
        self.__location = location

    def set_issue(self, issue):
        self.__issue = issue
    
    def get_name(self):
        return self.__name 

    def get_location(self):
        return self.__location

    def get_issue(self):
        return self.__issue


class Corporation:
    def __init__(self, name, location, sector):
        self.__name = name
        self.__location = location
        self.__sector = sector 

    def set_name(self, name):
        self.__name = name

    def set_location(self, location):
        self.__location = location

    def set_sector(self, sector):
        self.__sector = sector

    def get_name(self):
        return self.__name 

    def get_location(self):
        return self.__location

    def get_sector(self):
        return self.__sector



nonprofit = Nonprofit("North Texas Food Bank", "Texas", "Hunger")


print("Nonprofit Information:")
print("Name:", nonprofit.get_name())
print("Location:", nonprofit.get_location())
print("Issue:", nonprofit.get_issue())

corporation = Corporation("Central Market", "Texas", "Food/Retail")

print("\nCorporation Information:")
print("Name:", corporation.get_name())
print("Location:", corporation.get_location())
print("Sector:", corporation.get_sector())


queue = deque()

# Create a Nonprofit object and enqueue it
nonprofit = Nonprofit("North Texas Food Bank", "Texas", "Hunger")
queue.append(nonprofit)

# Create a Corporation object and enqueue it
corporation = Corporation("Central Market", "Texas", "Food/Retail")
queue.append(corporation)

# Print the information of each object in the queue
print("Queue Information:")
while queue:
    entity = queue.popleft()  # Dequeue the front object
    if isinstance(entity, Nonprofit):
        print("Nonprofit Information:")
        print("Name:", entity.get_name())
        print("Location:", entity.get_location())
        print("Issue:", entity.get_issue())
    elif isinstance(entity, Corporation):
        print("Corporation Information:")
        print("Name:", entity.get_name())
        print("Location:", entity.get_location())
        print("Sector:", entity.get_sector())
    print() 
```
